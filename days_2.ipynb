# Изображения

### Подключаем библиотеки

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("HWS.csv")
```

### Загрузить данные

```python
data = pd.read_csv("HWS.csv")
print(data.head())
print(data.shape)
```

### Гистограмма

```python
data["Height"].hist()
plt.show()
```

```python
data["Height"].hist(bins=25)
plt.show()
```

### Столбчатая диаграмма

```python
data_gr_height = data.groupby("Index")["Height"].mean()
print(data_gr_height)
```

```python
data_gr_height.plot(kind="bar", title="Средний рост по индексу массы тела")
plt.show()
```

### Загрузка датасета качество атмосферного воздуха

```python
aq = pd.read_csv("aq.csv")
print(aq)
```

### Линейный график

```python
aq.plot(x = "Date", y = "NO2", rot=45)
plt.show()
```

### Диаграмма рассеивания

```python
data.plot(x = "Height", y = "Weight", kind = "scatter")
plt.show()
```

### Добавление уровней

```python
data[data["Gender"]=="Male"]["Height"].hist(bins=25, alpha=0.7)
data[data["Gender"]=="Female"]["Height"].hist(bins=25, alpha=0.7)
plt.legend(["Male", "Female"])
plt.show()
```

## Объединение датафреймов

###  Подключение библиотек и загрузка данных

```python
import pickle
with open("census.p", "rb") as file:
    census = pickle.load(file)

cns = pd.DataFrame(census)
print(cns)

with open("ward.p", "rb") as file:
    ward = pickle.load(file)

wrd = pd.DataFrame(ward)
print(wrd)
```

### Обзор данных

```python
print(cns.shape)
print(cns.columns)
print(cns.head())
print(wrd.shape)
print(wrd.columns)
print(wrd.head())
```

### Объединения данных

```python
ward_census = wrd.merge(cns, on="ward")
print(ward_census.head(4))
print(ward_census.shape)
print(ward_census.columns)
```

```python
ward_census = wrd.merge(cns, on="ward", suffixes=("_ward", "_cen"))
print(ward_census.head(4))
```
